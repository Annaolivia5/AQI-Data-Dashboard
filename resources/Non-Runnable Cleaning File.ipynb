```python
import pandas as pd
import datetime as dt
```

```python
df = pd.read_csv('pollution_us_2000_2016.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['Date Local'] =  pd.to_datetime(df['Date Local'])
```

```python
df.info()
```

```python
df['Date Local'].dt.year
```

```python
poll_2016_df = df[df['Date Local'].dt.year == 2016]
#exclude = df[df['Date Local'].dt.year != 2016]
```

```python
poll_2016_df
```

```python
poll_2016_df.to_csv('Pollution_2016_csv.csv', index = False)
```

```python

pollution_2016_df = poll_2016_df[['Address', 'State', 'County', 'City', 'Date Local', 'NO2 AQI', 'O3 AQI', 'SO2 Mean', 'SO2 AQI', 'CO Mean','CO AQI']]
```

```python
pollution_2016_df
```

```python
pollution_2016_df = pollution_2016_df.fillna(0.0).copy()
```

```python
pollution_2016_df
```

```python
AQI_2016 = pollution_2016_df.groupby([pollution_2016_df['Date Local'].dt.month, 'State']).agg({'NO2 AQI': 'mean', 'O3 AQI': 'mean', 'SO2 AQI' :'mean', 'CO AQI': 'mean'})
```

```python
AQI_2016
```

```python
poll_2015_df = df[df['Date Local'].dt.year == 2015]
poll_2015_df 
```

```python
pollution_2015_df.to_csv('Pollution_2015.csv', index=False)
```

```python
poll_2015_df = poll_2015_df[['Address', 'State', 'County', 'City', 'Date Local', 'NO2 AQI', 'O3 AQI', 'SO2 Mean', 'SO2 AQI', 'CO Mean','CO AQI']]
```
